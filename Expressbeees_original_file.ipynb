```python
# Importing the dataset
dataset = pd.read_csv('Expressbees_csv.csv')
```

```python
chart = sns.countplot(
    x=dataset["lane"],data=dataset)

chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
plt.show()
```

```python
chart = sns.countplot(
    x=dataset["routemode"],data=dataset)

chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
plt.show()
```
